```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df["Cabin"] = df["Cabin"].fillna("Unknown")
```

```python
print("Total Passengers:", len(df))
```

```python
survivors = df["Survived"].sum()

print("Total Survivors:", survivors)
```

```python
survival_rate = (df["Survived"].mean()) * 100

print("Survival Rate:", round(survival_rate, 2), "%")
```

```python
print(df["Sex"].value_counts())
```

```python
print(df["Pclass"].value_counts())
```

```python
older_passengers = df[df["Age"] > 30]

print(older_passengers.head())
```

```python
female_survivors = df[
    (df["Sex"] == "female")
    &
    (df["Survived"] == 1)
]

print(female_survivors.head())
```

```python
gender_survival = df.groupby("Sex")["Survived"].mean()

print(gender_survival)
```

```python
avg_age = df.groupby("Sex")["Age"].mean()

print(avg_age)
```

```python
avg_fare = df.groupby("Pclass")["Fare"].mean()

print(avg_fare)
```

```python
highest_fare = df["Fare"].max()

print("Highest Fare:", highest_fare)
```

```python
df.to_csv("processed_titanic.csv", index=False)

print("Processed Dataset Saved Successfully")
```

```python
import matplotlib.pyplot as plt

df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Distribution")

plt.xlabel("Survived")

plt.ylabel("Count")

plt.show()
```
