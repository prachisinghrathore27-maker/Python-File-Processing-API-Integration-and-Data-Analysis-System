```python
import requests
```

```python
API_URL = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(API_URL)

if response.status_code == 200:
    data = response.json()
    print("Data fetched successfully!")
    print("Total records:", len(data))
else:
    print("Failed to fetch data")
```

```python
print("\n--- Sample Records ---")

for post in data[:5]:
    print(f"ID: {post['id']}")
    print(f"Title: {post['title']}")
    print("-" * 40)
```

```python
keyword = "qui"

print(f"\n--- Searching for '{keyword}' ---")

filtered_posts = []

for post in data:
    if keyword.lower() in post['title'].lower():
        filtered_posts.append(post)

for post in filtered_posts:
    print(f"ID: {post['id']}")
    print(f"Title: {post['title']}")
    print("-" * 40)

print("Total matches:", len(filtered_posts))
```

```python
print("\n--- Exception Handling Demo ---")

try:
    bad_url = "https://jsonplaceholder.typicode.com/invalid"

    response = requests.get(bad_url)

    response.raise_for_status()

except requests.exceptions.HTTPError as err:
    print("HTTP Error:", err)

except requests.exceptions.ConnectionError:
    print("Connection Error")

except requests.exceptions.Timeout:
    print("Request Timed Out")

except requests.exceptions.RequestException as err:
    print("Something went wrong:", err)
```

```python
def fetch_data(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.json()


def search_posts(data, keyword):
    results = []

    for item in data:
        if keyword.lower() in item['title'].lower():
            results.append(item)

    return results
```

```python
try:
    posts = fetch_data(API_URL)

    results = search_posts(posts, "qui")

    print("Matching Posts:", len(results))

except Exception as e:
    print("Error:", e)
```
