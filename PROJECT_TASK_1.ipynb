```python
import os
import shutil
import csv
```

```python
base_dir = "./file_demo"

if not os.path.exists(base_dir):
    os.mkdir(base_dir)
    print(f"Created directory: {base_dir}")
else:
    print("Directory already exists")
```

```python
text_file = os.path.join(base_dir, "sample.txt")

with open(text_file, "w") as file:
    file.write("Hello, this is a sample text file for automation demo.")

print("Sample text file created successfully.")
```

```python
print("\n--- Reading File ---")

with open(text_file, "r") as file:
    content = file.read()
    print("Content read:", content)
```

```python
csv_file = os.path.join(base_dir, "data.csv")

rows = [
    ["Task", "Status"],
    ["Task_A", "Completed"],
    ["Task_B", "Pending"]
]

with open(csv_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

print("CSV file created successfully.")
```

```python
print("\n--- Reading CSV Data ---")

with open(csv_file, "r") as file:
    reader = csv.reader(file)

    next(reader)  # Skip header

    for row in reader:
        print(f"Processing: {row[0]} - {row[1]}")
```

```python
new_text_file = os.path.join(base_dir, "archived_sample.txt")

os.rename(text_file, new_text_file)

print(f"Moved and Renamed '{text_file}' to '{new_text_file}'")
```

```python
print("\n--- Exception Handling Demo ---")

try:
    with open("missing_file.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("Error: File not found.")

except Exception as e:
    print("Unexpected Error:", e)
```

```python
delete_choice = False

if delete_choice:
    os.remove(new_text_file)
    print("Archived file deleted.")
```
